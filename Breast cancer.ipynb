```python
#  we transform the class labels from their original string representation('M' and 'B') into integers:
# using  a LabelEncoder object

X = Breast_cancer.iloc[:, 2:].values
y = Breast_cancer.iloc[:, 1].values
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
le.classes_
```

```python
le.transform(['M','B'])
```

```python
#  The random_state parameter,provides a fixed random seed
# the random_state=1 for the internal pseudo-random number generator that is used
# for shuffling the datasets prior to splitting. Using such a fixed random_state ensures
# that our results are reproducible.
# we take advantage of the built-in support for stratification via stratify=y. In
# this context, stratification means that the train_test_split method returns training
# and test subsets that have the same proportions of class labels as the input dataset. 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = \
train_test_split(X, y,test_size=0.20,stratify=y,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler#label encoding
from sklearn.decomposition import PCA#dimensionality reduction
from sklearn.linear_model import LogisticRegression#for estimation
from sklearn.pipeline import make_pipeline#chaiing transformations

```

```python
# cardinal rule of the pipeline.fit is that it takes an arbitrary number of scikit-learn transformers followed by a 
# scikit-learn estimator that implements the fit and predict
# transformers here been;standard scaler for label encoding and Principal Component Analysis(PCA)for dimensionality reduction
# estimator been Logistic regression to fit the data after it underwent transformation
# Long story short the estimator has to be the final element of the pipeline
pipe_lr = make_pipeline(StandardScaler(),PCA(n_components=2),LogisticRegression(random_state=1))
pipe_lr.fit(X_train, y_train)
```

```python
# at this point we are predicting the class label
y_pred = pipe_lr.predict(X_test)
print('Test Accuracy: %.3f' % pipe_lr.score(X_test, y_test))
```

```python
# am unsure of what happened so i wanted to test the model perfomance on data it has not seen
# Methodology:Stratified Kfold cross validation
from sklearn.model_selection import StratifiedKFold#intialise the iterator for better bias variace trade off
kfold = StratifiedKFold(n_splits=10,random_state=1).split(X_train, y_train)#there are 10 folds as specified in the n_splits
scores = [] #start with an empty list and assign it to the variable scores 
for k, (train, test) in enumerate(kfold):#we need a count in the folds so we use the enumerate function 
    pipe_lr.fit(X_train[train], y_train[train])#just to ensure that there is proper scaling in each iteration
    score = pipe_lr.score(X_train[test], y_train[test])#use the test indices to calculate the accuracy of the model
    scores.append(score)#add elements to the list 
    print('Fold: %2d, Class dist.: %s, Acc: %.3f' % (k+1,np.bincount(y_train[train]), score))#bincout to count the occurence of each element.
```

```python
 print('\nCV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores))) #average outcome of the scores 
```

```python
 from sklearn.model_selection import cross_val_score#distribute the evaluation of the different folds across multiple CPUs on our machine
scores = cross_val_score(estimator=pipe_lr,X=X_train,y=y_train,cv=10,n_jobs=1)#only one cpu will be used since n_jobs=1
print('CV accuracy scores: %s' % scores)
```

```python
from sklearn.model_selection import GridSearchCV  #for the use to tune a support vector machine
from sklearn.svm import SVC
pipe_svc = make_pipeline(StandardScaler(),
                         SVC(random_state=1))
param_range= [0.0001, 0.001, 0.01, 0.1,1.0, 10.0, 100.0, 1000.0]
param_grid = [{'svc__C': param_range,'svc__kernel': ['linear']},{'svc__C': param_range,'svc__gamma': param_range,
                                                                 'svc__kernel': ['rbf']}]#tune
#both the svc__C and svc__gamma parameter. Note that the svc__gamma parameter is
# specific to kernel SVMs.
gs = GridSearchCV(estimator=pipe_svc,param_grid=param_grid,
                  scoring='accuracy',cv=10,n_jobs=-1)#train data to perform grid search
gs = gs.fit(X_train, y_train)
print('Best score:', gs.best_score_)#we obtained the score of the best-performing model via the best_score_
print('Best parameters:',gs.best_params_)
```

```python
#  we will use the independent test dataset to estimate the performance of the
# best-selected model, which is available via the best_estimator_ attribute of the
# GridSearchCV object:
clf = gs.best_estimator_
clf.fit(X_train, y_train)
print('Test accuracy: %.3f' % clf.score(X_test, y_test))
```

```python
# performing a 5x2 nested cross validation i.e with five outer test folds and two inner train folds
# the returned cross validation score gives a good estimate of what to expect
# if we tune the hyperparameter of a model and use it on unseen data
gs = GridSearchCV(estimator=pipe_svc,param_grid=param_grid,scoring='accuracy',cv=2)
scores = cross_val_score(gs, X_train, y_train,scoring='accuracy', cv=5)
print('CV accuracy: %.3f +/- %.3f' % (np.mean(scores),np.std(scores)))
```

```python
# to select models for classification based on their 
# to test performance with respect to theTrue positive rate (TPR) and False positive rate (FPR) are performance metrics
# that are especially useful for imbalanced class problems 
# which are computed by shifting the decision threshold of the classifier
# A perfect classifier would fall into the top left corner of the graph with a TPR of 1 and an FPR of 0
#  ROC Area Under the Curve (ROC AUC) to characterize the performance of a classification model
from sklearn.metrics import roc_curve, auc 
from scipy import interp#one dimensional linear interpolation into three folds
pipe_lr = make_pipeline(StandardScaler(),PCA(n_components=2),LogisticRegression(penalty='l2'aper,random_state=1,C=100.0))
X_train2 = X_train[:, [4, 14]]
cv = list(StratifiedKFold(n_splits=3,random_state=1).split(X_train,y_train))
fig = plt.figure(figsize=(7, 5))
mean_tpr = 0.0#setting mean of the True positive rate to 0
mean_fpr = np.linspace(0, 1, 100)
all_tpr = []
for i, (train, test) in enumerate(cv):
    probas = pipe_lr.fit(X_train2[train],y_train[train]).predict_proba(X_train2[test])
    fpr, tpr, thresholds = roc_curve(y_train[test],probas[:, 1],pos_label=1)#for all rows and one column
mean_tpr += interp(mean_fpr, fpr, tpr)
mean_tpr[0] = 0.0
roc_auc = auc(fpr, tpr) #reciever operating characteristic area under curve 
# which the area under the curve for True positive rate (TPR) and False positive rate (FPR) are performance metrics
# that are especially useful for imbalanced class problems
plt.plot(fpr,tpr,label='ROC fold %d (area = %0.2f)'% (i+1, roc_auc))
plt.plot([0, 1],[0, 1],linestyle='--',color=(0.6, 0.6, 0.6),label='random guessing')
mean_tpr /= len(cv)
mean_tpr[-1] = 1.0
mean_auc = auc(mean_fpr, mean_tpr)
plt.plot(mean_fpr, mean_tpr, 'k--',label='mean ROC (area = %0.2f)' % mean_auc, lw=2)
plt.plot([0, 0, 1],[0, 1, 1],linestyle=':',color='black',label='perfect performance')
plt.xlim([-0.05, 1.05])#setting top and bottom limits for x-axis
plt.ylim([-0.05, 1.05]) #setting top and bottom limits for y-axis
plt.xlabel('false positive rate') #labelling the x-axis
plt.ylabel('true positive rate')#labelling the y-axis
plt.legend(loc="lower right")#placing the legend on the lower right
plt.show()
```

```python
#  let's create an imbalanced dataset from our breast cancer dataset, which
# originally consisted of 357 benign tumors (class 0) and 212 malignant tumors (class 1)
# take 357 benign tumor samples and stack
# them with the first 40 malignant samples to create a stark class imbalance
X_imb = np.vstack((X[y == 0], X[y == 1][:40]))
y_imb = np.hstack((y[y == 0], y[y == 1][:40]))
```

```python
# we would have an approximately 90% prediction accuracy  having used the above stack method
y_pred = np.zeros(y_imb.shape[0])
np.mean(y_pred == y_imb) * 100
```

```python
#upsampling of the minority class by drawing new samples from the dataset with replacement
from sklearn.utils import resample 
print('Number of class 1 samples before:',X_imb[y_imb == 1].shape[0])
X_upsampled, y_upsampled = resample(X_imb[y_imb == 1],y_imb[y_imb == 1],
                                    replace=True,n_samples=X_imb[y_imb == 0].shape[0],random_state=123)
print('Number of class 1 samples after:',X_upsampled.shape[0])
```

```python
#After resampling stack the original class 0 samples with the upsampled
# class 1 subset to obtain a balanced dataset 
X_bal = np.vstack((X[y == 0], X_upsampled))
y_bal = np.hstack((y[y == 0], y_upsampled))
```

```python
#  a majority vote prediction rule would only achieve 50 percent
# accuracy
y_pred = np.zeros(y_bal.shape[0])
np.mean(y_pred == y_bal) * 100
```
